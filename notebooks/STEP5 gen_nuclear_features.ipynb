```python
import os
import ast
import pickle
import openslide
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

nuc_features = [
    "density",
    "mean nucleus area",
    "anisocaryose",
    "nucleocyto index",
]
df = pd.DataFrame(columns=["lame"] + nuc_features)
df["lame"] = [e.split("_")[0] for e in os.listdir("../checkpoints/nucleus_checkpoints")]
```

```python
for data_name in os.listdir("../checkpoints/nucleus_checkpoints"):
    data = pd.read_csv("../checkpoints/nucleus_checkpoints/" + data_name)
    slide_name = data_name.split("_")[0]
    df.loc[df["lame"] == slide_name] = [slide_name] + list(data[nuc_features].mean())

df["patient"] = df["lame"].apply(lambda x: x[:-1]).astype(int)
```

```python
df_nuclear = pd.DataFrame(
    index=df["patient"].unique(),
    columns=["patient"] + nuc_features,
)

df_nuclear["patient"] = df["patient"].unique()
for patient in df["patient"].unique():
    df_nuclear.loc[df_nuclear["patient"] == patient] = [patient] + list(
        df.loc[df["patient"] == patient][nuc_features].mean()
    )

df_nuclear.head()
```

```python
df_nuclear.to_csv("../data/tabs/final_nuclear_features.csv", index=False)
```
