```python
import os
import matplotlib.pyplot as plt

from utils.utils_tumor import color_transfer
```

```python
reference_pb = plt.imread("HES__5.jpeg")

target_hm = plt.imread(
    "../data/patches/111B/" + os.listdir("../data/patches/111B")[1500]
)
target_bj = plt.imread(
    "../data/patches/121C/" + os.listdir("../data/patches/121C")[1500]
)

normalized_hm = color_transfer(target_hm, reference_pb)
normalized_bj = color_transfer(target_bj, reference_pb)
```

```python
fig, ax = plt.subplots(ncols=3, nrows=1, figsize=(9,3))
ax[0].imshow(reference_pb)
ax[1].imshow(target_hm)
ax[2].imshow(normalized_hm)
plt.tight_layout()
plt.show()
fig, ax = plt.subplots(ncols=3, nrows=1, figsize=(9,3))
ax[0].imshow(reference_pb)
ax[1].imshow(target_bj)
ax[2].imshow(normalized_bj)
plt.tight_layout()
plt.show()
```
