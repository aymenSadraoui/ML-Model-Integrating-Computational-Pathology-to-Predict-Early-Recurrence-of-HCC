### This operation is to resize the patches for inflammatory cells detection to reduce the computation time

```python
import os
from PIL import Image
from tqdm import tqdm

if not os.path.exists("../data/patches_bis"):
    os.mkdir("../data/patches_bis")
```

```python
for patches in os.listdir("../data/patches"):
    print(patches)
    for patch_dir in tqdm(os.listdir("../data/patches/" + patches)):
        if not os.path.exists("../data/patches_bis/" + patches):
            os.mkdir("../data/patches_bis/" + patches)
        img = Image.open("../data/patches/" + patches + "/" + patch_dir)
        if not os.path.exists("../data/patches_bis/" + patches + "/" + patch_dir):
            if img.size == (1152, 1152):
                img.resize((512, 512)).save(
                    "../data/patches_bis/" + patches + "/" + patch_dir
                )
            elif img.size == (626, 626):
                img.resize((278, 278)).save(
                    "../data/patches_bis/" + patches + "/" + patch_dir
                )
            else:
                img.resize((486, 486)).save(
                    "../data/patches_bis/" + patches + "/" + patch_dir
                )
```
